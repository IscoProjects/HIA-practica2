# Regresión Lineal y Regresión Logística

## Práctica: Predicción
Mediante la técnica de Webscraping se obtuvo una base de datos que contiene información sobre artículos de Machine Learning. Las características de entrada en el archivo son:
* Title: título del artículo
* url: ruta al artículo
* Word count: número de palabras del artículo
* '#' of links: enlaces externos del artículo
* '#' of comments: número de comentarios
* '#' of Images video: imágenes o videos que contiene el artículo
* Elapsed days: días transcurridos desde que se creó el artículo
* '#' shares: veces que se compartío el artículo

## Regresión Lineal Simple

Por lo tanto, a partir de las características de los artículos de ML el objetivo de esta práctica es predecir las veces que será compartido un artículo

```python
#Importar las librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
```

```python
#Leer los datos
df = pd.read_csv('ml_articles_data.csv', sep = ';')
df.head(10)
```

```python
df.describe()
```

Se observa que la media de palabras en los artículos es 1808, el mínimo es 250 y el máximo es 8401. Mediante la relación lineal, intentaremos identificar si existe una correlación entre la cantidad de palabras del artículo y el número de veces que es compartido

```python
#Visualizar las características de interés
df.drop(['Title', 'url', 'Elapsed days'], axis = 1).hist()
plt.show()
```

Filtramos el número de palabras y el número de compartidos para trabajar con los valores donde se concentran la mayoría de registros

```python
#Seleccionar la zona de datos concentrada
#Eje x: Word count ente 0 y 3500
#Eje y: Shares entre 0 y 80 000
filtered_data = df[(df['Word count'] <= 3500) & (df['# Shares'] <= 80000)]
colors = ['orange', 'blue']
size = [30,60]
#f1 y f2 son arreglos que contienen los valores de las columnas filtradas.
f1 = filtered_data['Word count'].values
f2 = filtered_data['# Shares'].values
```

```python
f1
```

```python
f2
```

Graficamos 

```python
#Inicializar una lista vacia
assign = []
#Iterar sobre cada fila del DataFrame filtered_data.
for index, row in filtered_data.iterrows():
    if(row['Word count'] > 1808):
        #Puntos de color naranja si el conteo de palabras es mayor que 1808
        assign.append(colors[0])
    else:
        #Puntos de color azul si el conteo de palabras es menor o igual que 1808
        assign.append(colors[1])
#Crear un gráfico de dispersión (scatter plot) 
#f1 y f2 para los ejes x e y, respectivamente.
plt.scatter(f1, f2, c = assign, s = size[0])
plt.show()
```

Para la práctica de regresión lineal simple, los datos de entrada son Word Count y # of Shares

```python
#Crear un objeto de regresión lineal
reg = linear_model.LinearRegression()
#Asignar x como variable de entrada para el entrenamiento
data_x = filtered_data['Word count']
#Conversión a un Arreglo
x_train = np.array(data_x).reshape(-1, 1)
#x_train = filtered_data[['Word count']]
#Selección de la Variable de Salida 
y_train = filtered_data['# Shares'].values
#Se ajusta el modelo de regresión lineal a los datos de entrenamiento
reg.fit(x_train, y_train)
```

Predicciones

```python
#Se hacen predicciones utilizando el modelo ajustado
y_pred = reg.predict(x_train)
```

Análisis de los resultados

```python
#Devolver el coeficiente de la regresión lineal (pendiente de la recta ajustada).
reg_coef = reg.coef_
#Devolver el intercepto de la regresión lineal (ordenada al origen de la recta ajustada)
reg_intr = reg.intercept_
# El MSE mide la media de los cuadrados de los errores, es decir, la diferencia promedio entre los valores predichos y los valores reales
mse = mean_squared_error(y_train, y_pred)
#R² indica la proporción de la variabilidad en los datos que es explicada por el modelo
variance = r2_score(y_train, y_pred)

#mostrar los resultados
print('Coeficiente:', reg_coef)
print('Termino Independiente:', reg_intr)
print('MSE:', mse)
print('Varianza: (R2)', variance)
```

Visualizar la recta

```python
#Mostrar la grafica
plt.scatter(x_train, y_train, c=assign, s=30)
plt.plot(x_train, y_pred, color='red', linewidth=2)
plt.xlabel('Cantidad de Palabras')
plt.ylabel('Compartido en Redes')
plt.title('Regresión Lineal')

# Ajustar las posiciones de las métricas en la gráfica
x_pos = min(x_train) + (max(x_train) - min(x_train)) * 0.05  # 5% de margen a la derecha del mínimo de x
y_pos_start = max(y_train) - (max(y_train) - min(y_train)) * 0.05  # 5% desde el máximo de y

# Mostrar las métricas en la gráfica con espaciado entre líneas
plt.text(x_pos, y_pos_start, f'Coeficiente: {reg_coef[0]:.2f}', fontsize=12, color='black')
plt.text(x_pos, y_pos_start - 5000, f'Intercepto: {reg_intr:.2f}', fontsize=12, color='black')
plt.text(x_pos, y_pos_start - 10000, f'MSE: {mse:.2f}', fontsize=12, color='black')
plt.text(x_pos, y_pos_start - 15000, f'R²: {variance:.2f}', fontsize=12, color='black')

plt.show()
```

Queremos predecir cuántas veces se compartiría un artículo de ML basado en el número de palabras. Por ejemplo, un artículo de 2500 palabras.

```python
#Predecir el número de shares que tendrá un articulo con x número de palabras
y_words = reg.predict([[2500]])
response = int(y_words[0])
print('Prediccion Nro. de compartidas:', response)
```

## Regresión Lineal Múltiple

Ahora, utilizaremos 2 variablespredictoras. La primera seguirá siendo Word count, y la segunda será una suma de las 3 columnas de entrada: número de enlaces, número de comentarios y número de imágenes o video

```python
filtered_data
```

```python
#Creamos una nueva variable que sera la suma de # of links, # of comments y # of images/video
sum = (filtered_data['# of Links'] + filtered_data['# of comments'].fillna(0) + filtered_data['# Images video'])

data_x = pd.DataFrame()
data_x['Word count'] = filtered_data['Word count']
data_x['Sum'] = sum
data_x
```

Crear el modelo de regresión y entrenar el modelo

```python
#Definir los datos de entrenamiento
xy_train = np.array(data_x)
z_train = filtered_data['# Shares']

#Crear un objeto de regresion lineal
reg_mul = linear_model.LinearRegression()

#Entrenar el modelo con las dos dimensiones xy
reg_mul.fit(xy_train, z_train)
```

Predecir

```python
#Predicción
z_pred = reg_mul.predict(xy_train)
```

Análisis de los resultados

```python
#Devolver el coeficiente de la regresión lineal (pendiente de la recta ajustada).
reg_coef = reg_mul.coef_
# El MSE mide la media de los cuadrados de los errores, es decir, la diferencia promedio entre los valores predichos y los valores reales
mse = mean_squared_error(z_train, z_pred)
#R² indica la proporción de la variabilidad en los datos que es explicada por el modelo
variance = r2_score(z_train, z_pred)

#mostrar los resultados
print('Coeficiente:', reg_coef)
print('MSE:', mse)
print('Varianza:', variance)
```

Visualizar

```python
#Graficar un plano de tres dimensiones
fig = plt.figure()
fig = plt.figure(figsize=(16, 16)) 
ax = fig.add_subplot(111, projection='3d')

#Crear una malla sobre la que graficaremos el plano
xx, yy = np.meshgrid(np.linspace(0, 3500, num = 10), np.linspace(0, 60, num = 10))

#Calcular los valores del plano para los puntos x e y
new_x = (reg_mul.coef_[0] * xx)
new_y = (reg_mul.coef_[1] * yy)

#Calcular los correspondientes valores para z. Sumar el punto de intercepci+on
z = (new_x + new_y + reg_mul.intercept_)

#Graficamos el plano
ax.plot_surface(xx, yy, z, alpha = 0.2, cmap = 'hot')
#Graficamos en azul los puntos en 3D
ax.scatter(xy_train[:, 0], xy_train[:, 1], z_train, c='blue',s=30)
ax.scatter(xy_train[:, 0], xy_train[:, 1], z_pred, c='red',s=40)

ax.view_init(elev=30., azim=65)

ax.set_xlabel('Cantidad de Palabras')
ax.set_ylabel('Cantidad de Enlaces,Comentarios e Imagenes')
ax.set_zlabel('Compartido en Redes')
ax.set_title('Regresión Lineal con Múltiples Variables')

plt.show()
```

Queremos predecir cuántas veces se compartiría un artículo de ML basado en el número de palabras, enlaces, comentarios e imágenes. Por ejemplo, un artículo de 2500 palabras, con 5 enlaces, 10 comentarios y 5 imágenes

```python
#Predecir cuantos shares tendrá un articulos insertando el numero de palabras, el numero de enlaces, el numero de comentarios y el numero de enlaces
y_words = reg_mul.predict([[2500, 5+10+5]])
response = int(y_words[0])
print('Prediccion Nro. de Compartidas:', response)
```
